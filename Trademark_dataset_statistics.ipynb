```python
import matplotlib.pyplot as plt

# Filtered data
datasets = [
    "BelgaLogos",
    "FlickrLogos",
    "MICC-Logos",
    "Euro 2008"
]

years = [2009, 2010, 2013, 2010]
image_counts = [10000, 8240, 720, 106]

# Combine and sort by year
combined = sorted(zip(years, datasets, image_counts))
sorted_years, sorted_datasets, sorted_counts = zip(*combined)

# Plot
plt.figure(figsize=(10, 5))
bars = plt.bar(sorted_datasets, sorted_counts, color='mediumseagreen')

# Add image count labels
for bar, count in zip(bars, sorted_counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{count:,}",
             ha='center', va='bottom', fontsize=10)

# Year labels under each bar
for i, year in enumerate(sorted_years):
    plt.text(i, -max(sorted_counts) * 0.05, f"Year: {year}", ha='center', va='top', fontsize=9)

# Aesthetics
plt.title("Selected Logo Datasets Over Time", fontsize=14)
plt.xlabel("Dataset", fontsize=12)
plt.ylabel("Number of Images", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(bottom=-max(sorted_counts) * 0.1)

plt.tight_layout()
plt.show()
```
